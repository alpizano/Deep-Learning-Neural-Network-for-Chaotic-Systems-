```python
def imshow(inp):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
```

```python
#simple_transform = transforms.Compose([transforms.Resize((224,224))
                                       #,transforms.ToTensor()
                                      # ,transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
simple_transform = transforms.Compose([transforms.Resize((32,32))
                                       ,transforms.ToTensor()
                                       ,transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
train = ImageFolder('dogsandcats/train/train',simple_transform)
valid = ImageFolder('dogsandcats/train/valid/',simple_transform)

print (train) # train folder has 23,000 unique images
print (valid) # valid has 2,000 unique seperated into dog & cat folder not found in train.
```

```python
# Convert "train" and "valid" datasets into dataloaders
train_data_gen = torch.utils.data.DataLoader(train,shuffle=True,batch_size=64,num_workers=3)
valid_data_gen = torch.utils.data.DataLoader(valid,batch_size=64,num_workers=3)
```

```python
dataset_sizes = {'train':len(train_data_gen.dataset),'valid':len(valid_data_gen.dataset)}
```

```python
dataloaders = {'train':train_data_gen,'valid':valid_data_gen}
```

```python
# Defining the CNN from scratch
import torch.nn.functional as F

# pass nn.Module as parameter to subclass parent class (inheritance in python)
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5) # 3 channel input img, 6 feautures output, 5x5 sqaure convolution
        self.pool = nn.MaxPool2d(2, 2) # 2x2 max pooling window, stride 1
        self.conv2 = nn.Conv2d(6, 16, 5) # 6 feature input from above, 16 feautures output, 5x5 sqaure convolution
        # an affine operation: y = Wx + b
        self.fc1 = nn.Linear(16 * 5 * 5, 120) #16 5x5 inputs fully connecting in linear layer 1
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        #x = x.view(-1, self.num_flat_features(x))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
net = Net()

print (net)
```

```python
# defining loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
# Train the network over 2 iterations

for epoch in range(2):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(train_data_gen, 0):
        # get the inputs
        inputs, labels = data
       # print ("The input is: %s" % inputs)
        # print ("The label is: %s" % labels)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        # perform backward pass
        loss.backward()
        # update the weights
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        #if i % 2000 == 1999:    # print every 2000 mini-batches
        #print('[%d, %5d] loss: %.3f' % (epoch + 1, i + 1, running_loss / 2000))
        print('[%d, %5d] loss: %.3f' % (epoch + 1, i + 1, running_loss))
        running_loss = 0.0

print('Finished Training')
```
